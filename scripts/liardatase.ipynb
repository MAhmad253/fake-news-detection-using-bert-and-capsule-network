```python
def convert_label(x):
    if x in [0, 1]:        # pants-fire, false
        return 0  # False
    elif x in [4, 5]:      # mostly-true, true
        return 1  # True
    else:                  # barely-true, half-true
        return 2  # Misleading

filtered_df['label'] = filtered_df['label'].apply(convert_label)
```

```python
def map_label(row):
    if row['True'] == 1.0 and row['False'] == 0.0:
        return 1  # True
    elif row['True'] == 0.0 and row['False'] == 1.0:
        return 0  # False
    elif row['True'] == 0.5 and row['False'] == 0.5:
        return 2  # Misleading
    else:
        return None  # unexpected case

# Apply row-wise
merge_df['label'] = merge_df.apply(map_label, axis=1)

# Check result
print(merge_df['label'].value_counts())
```

```python
import pandas as pd
df=pd.read_csv("../data/combinedfully_liar.csv")
```

```python
df.columns
```

```python
import pandas as pd

def split_statement(text):
    words = text.split()
    
    # First 12 words → title
    title = ' '.join(words[:12])
    
    # Full text → text
    return pd.Series([title, text])

# Apply OUTSIDE the function
df[['title', 'text']] = df['statement'].apply(split_statement)
```

```python
df.head(10)
```

```python
df1=df.drop("statement",axis=1)
```

```python
# Reorder columns: statement → text → label
df2 = df1[['title', 'text', 'label']]

df2.head(10)
```

```python
print("Duplicate titles:", df2['title'].duplicated().sum())
```

```python
# Remove all duplicate rows based on 'statement'
df3 = df2.drop_duplicates(subset=['title']).reset_index(drop=True)

# Verify
print("Remaining rows:", df3.shape)
print("Duplicate titles after removal:", df3['title'].duplicated().sum())
```

```python
df3.to_csv("../data/combinedfully_liar123.csv",index=False)
```
