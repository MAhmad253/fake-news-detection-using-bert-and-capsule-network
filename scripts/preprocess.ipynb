```python
import re

def is_corrupted(text):                         # very important pipeline for suspecious data. this will remove the corrupted data, use this cell next and then next
    if not isinstance(text, str):
        return False

    # 1. Typical mojibake patterns
    if re.search(r'Ã.|Â.|â.', text):
        return True

    # 2. Too many suspicious characters
    suspicious = sum(1 for c in text if ord(c) in range(128, 256))
    if suspicious > len(text) * 0.2:  # >20% suspicious chars
        return True

    # 3. Repeated weird sequences like Ã°Â
    if re.search(r'(Ã.|Â.|â.){2,}', text):
        return True

    return False
```

```python
def fix_mojibake(text):
    if not isinstance(text, str):
        return text
    
    for _ in range(5):  # multiple passes for heavy corruption
        try:
            new_text = text.encode('latin1').decode('utf-8')
            if new_text == text:
                break
            text = new_text
        except:
            break
    return text
```

```python
for col in df.select_dtypes(include='object').columns:
    df7[col] = df7[col].apply(fix_mojibake)
```

```python
import pandas as pd
df=pd.read_csv("../data/previousfinal.csv")
```

```python
import re

def clean_title(text):
    if pd.isna(text):
        return text

    text = text.lower()
    
    # Remove abusive words (basic)
    text = re.sub(r'f\*cking|stupid|bitch|lunatic', '', text)
    
    # Fix spacing issues
    text = re.sub(r'([a-z])([A-Z])', r'\1 \2', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['title'] = df['title'].apply(clean_title)
```

```python
import re

def clean_title(text):
    if pd.isna(text):
        return text
    
    # Remove [video], [VIDEO], etc.
    text = re.sub(r'\[.*?\]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['title'] = df['title'].apply(clean_title)
```

```python
import re

def clean_brackets(text):
    if pd.isna(text):
        return text
    
    # Remove only useless brackets
    text = re.sub(r'\((video|details|update|tweet|photo)\)', '', text, flags=re.IGNORECASE)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['title'] = df['title'].apply(clean_brackets)
```

```python
import re

def remove_small_brackets(text):
    if pd.isna(text):
        return text
    
    # Remove brackets where inside length < 6 characters
    text = re.sub(r'\([^)]{0,5}\)', '', text)
    
    # Clean extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['title'] = df['title'].apply(remove_small_brackets)
```

```python
import re

# Remove (tweets) specifically (case-insensitive)
df['title'] = df['title'].str.replace(r'\(tweets\)', '', regex=True, flags=re.IGNORECASE)

# Clean extra spaces
df['title'] = df['title'].str.replace(r'\s+', ' ', regex=True).str.strip()
```
