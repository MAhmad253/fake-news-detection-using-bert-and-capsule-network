```python
import re

def clean_final(text):
    if pd.isna(text):
        return text

    text = text.lower()

    # ❌ Remove "read more"
    text = re.sub(r'read more', '', text)

    # ❌ Remove URLs
    text = re.sub(r'http\S+', '', text)

    # ❌ Remove multiple dots (...)
    text = re.sub(r'\.{2,}', '.', text)

    # ❌ Remove multiple commas
    text = re.sub(r',+', ',', text)

    # ❌ Remove repeated punctuation (!! ??)
    text = re.sub(r'([!?]){2,}', r'\1', text)

    # ❌ Remove unwanted characters (keep basic punctuation)
    text = re.sub(r'[^a-z0-9\s\.\,\!\?%]', '', text)

    # ❌ Fix spacing around punctuation
    text = re.sub(r'\s*([.,!?])\s*', r'\1 ', text)

    # ❌ Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
df6['title'] = df6['title'].apply(clean_final)
df6['text'] = df6['text'].apply(clean_final)
```

```python
import re

def extract_misleading(text):
    if pd.isna(text):
        return None
    
    text = text.lower()

    # ❌ Remove metadata
    text = re.sub(r'written by.*?\d{4}', '', text)
    
    # ❌ Remove "read more"
    text = re.sub(r'read more', '', text)

    # ❌ Remove negation words
    text = re.sub(r"\b(no|not|doesnt|didnt|false|isnt|arent|never)\b", '', text)

    # ❌ Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# =========================
# APPLY ON DATASET
# =========================

# If your column is 'text' or 'statement'
df6['text'] = df6['text'].apply(extract_misleading)

# Remove empty rows
df7 = df6.dropna()

# Remove very short rows (important)
df8 = df7[df7['text'].str.len() > 20]

# =========================
# SAVE
# =========================
# df.to_csv("misleading_converted.csv", index=False)
df8.to_csv("../data/finaldatasets/misleadingfile2.csv", index=False, encoding="utf-8-sig")

print("Done ✅", df.shape)
```
