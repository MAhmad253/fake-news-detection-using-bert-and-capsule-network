```python
import re

def normalize_text(text: str):
    # Fix quotes
    text = text.replace("“", '"').replace("”", '"').replace("‘", "'").replace("’", "'")

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r"http\S+", "", text)

    # Keep important punctuation (! ? .)
    text = re.sub(r"[^a-zA-Z0-9\s\.\!\?]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text


def shorten_text(text: str, max_words=60):
    words = text.split()
    return " ".join(words[:max_words])


# def create_input(title: str, text: str):
#     title = "" if title is None else title
#     text = "" if text is None else text

#     title = normalize_text(title)
#     text = normalize_text(text)

#     if text.strip() == "":
#         return title

#     short_text = shorten_text(text)
#     return title + " " + short_text


# def preprocess_pipeline(title: str, text: str):
#     input_text = create_input(title, text)

#     return {
#         "input_text": input_text
#     }
# Apply normalization
final_df['title'] = final_df['title'].fillna('').apply(normalize_text)
final_df['text'] = final_df['text'].fillna('').apply(normalize_text)

# Shorten text
final_df['text'] = final_df['text'].apply(shorten_text)


# Reset index
final_df = final_df.reset_index(drop=True)

# Check
final_df.head(30)
# final_df.to_csv("../data/finaldatasets/completedata.csv",index=False)
# # =========================
# # TEST
# # =========================
# title = "BREAKING!!! Pakistan wins the T20 World Cup 2026!!!"
# text = "Pakistan has officially won the T20 World Cup after defeating India in a thrilling final match."

# print(preprocess_pipeline(title, text))
```
