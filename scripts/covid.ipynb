```python
from ftfy import fix_text

df2["title"] = df2["title"].astype(str).apply(fix_text)
df2["content"] = df2["content"].astype(str).apply(fix_text)
```

```python
df3["title"] = df3["title"].str.lower().str.strip()
df4 = df3.drop_duplicates(subset=["title"])
```

```python
df4["content"] = df4["content"].str.lower().str.strip()
df5 = df4.drop_duplicates(subset=["content"])
```

```python
print(df5.duplicated().sum())
```

```python
import re

def clean_text(text):
    if pd.isna(text):
        return text
    
    text = text.lower()
    
    # Fix common errors
    text = text.replace("covid-lg", "covid-19")
    
    # Remove "getty images"
    text = re.sub(r'getty images', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df5['content'] = df5['content'].apply(clean_text)
```

```python
final_df = df5[df5['content'].str.len() > 50]
```

```python
import re

def remove_prefix(text):
    if pd.isna(text):
        return text
    
    text = re.sub(r'^.*?coronavirus news\.?', '', text)
    return text.strip()

final_df['content'] = final_df['content'].apply(remove_prefix)
```

```python
import re

def clean_title(text):
    if pd.isna(text):
        return text
    
    text = text.lower()
    
    # Fix common spacing issues
    text = re.sub(r'([a-z])([A-Z])', r'\1 \2', text)  # camel case
    text = re.sub(r'(?<=\w)(?=where|finds|you|the)', ' ', text)
    
    # Fix common typos
    text = text.replace("doesnt", "doesn't")
    text = text.replace("dont", "don't")
    text = text.replace("arent", "aren't")
    text = text.replace("dotft", "don't")
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

final_df['title'] = final_df['title'].apply(clean_title)
```

```python
import re
def shorten_text(text):
    if pd.isna(text):
        return text
    sentences = re.split(r'(?<=[.!?]) +', text)
    short = ' '.join(sentences[:3])
    return ' '.join(short.split()[:60])  # max 60 words
final_df['content'] = final_df['content'].apply(shorten_text)
```

```python
import pandas as pd

def split_statement(text):
    words = text.split()
    
    # First 12 words → title
    title = ' '.join(words[:12])
    
    # Full text → text
    return pd.Series([title, text])

# Apply OUTSIDE the function
df1[['title', 'statement']] = df1['text'].apply(split_statement)
```

```python
df2=df1.drop("text",axis=1)
df2.head(10)
```

```python
df4 = df3[['title', 'text', 'label']]

df4.head(10)
```
